# Life Expectancy and GDP

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the CSV file 
df = pd.read_csv('all_data.csv')
```

```python
#Changing 'Life expectancy' column name to 'LEB'
df.rename(columns={'Life expectancy at birth (years)': 'LEB'}, inplace=True)
```

```python
#Making a function that calculates the mean of LEB for each country 
def leb_mean(country, data):
    subset = data[data.Country == country]
    mean = np.mean(subset.LEB)
    return '{} LEB mean (years 2000-2015): {}'.format(country, mean)

chile_leb_mean = leb_mean('Chile', df)
china_leb_mean = leb_mean('China', df)
germany_leb_mean = leb_mean('Germany', df)
mexico_leb_mean = leb_mean('Mexico', df)
usa_leb_mean = leb_mean('United States of America', df)
zimbabwe_leb_mean = leb_mean('Zimbabwe', df)

print(chile_leb_mean)
print(china_leb_mean)
print(germany_leb_mean)
print(mexico_leb_mean)
print(usa_leb_mean)
print(zimbabwe_leb_mean)
```

```python
#Between years 2000 and 2015, Germany had the highest LEB mean. Zimbabwe is the last in the list, 
#with almost 30 years less than Germany
```

```python
#Plotting line graphs of the LEB evolution over 15 years for each country
chile_line = df[df.Country == 'Chile']
china_line = df[df.Country == 'China']
germany_line = df[df.Country == 'Germany']
mexico_line = df[df.Country == 'Mexico']
usa_line = df[df.Country == 'United States of America']
zimbabwe_line = df[df.Country == 'Zimbabwe']

years = ['2000', '2005', '2010', '2015']

plt.figure(figsize=(10, 7))
ax = plt.subplot()
plt.plot(chile_line.Year, chile_line.LEB, color='gold', label='Chile')
plt.plot(china_line.Year, china_line.LEB, color='red', label='China')
plt.plot(germany_line.Year, germany_line.LEB, color='blue', label='Germany')
plt.plot(mexico_line.Year, mexico_line.LEB, color='purple', label='Mexico')
plt.plot(usa_line.Year, usa_line.LEB, color='green', label='USA')
plt.plot(zimbabwe_line.Year, zimbabwe_line.LEB, color='magenta', label='Zimbabwe')
ax.set_xticks(range(2000, 2016)) 
ax.set_xticklabels(['2000', '2001', '2002', '2003', '2004', '2005', '2006', 
                   '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015'], rotation=30, fontsize=10)
plt.xlabel('Years')
plt.ylabel('LEB')
plt.title('LEB evolution over 15 years')
plt.legend()
plt.show()

plt.figure(figsize=(17, 10))

ax = plt.subplot(2, 3, 1)
plt.plot(chile_line.Year, chile_line.LEB, color='gold')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('LEB')
plt.title('Chile')

ax = plt.subplot(2, 3, 2)
plt.plot(china_line.Year, china_line.LEB, color='red')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('LEB')
plt.title('China')

ax = plt.subplot(2, 3, 3)
plt.plot(germany_line.Year, germany_line.LEB, color='blue')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('LEB')
plt.title('Germany')

ax = plt.subplot(2, 3, 4)
plt.plot(mexico_line.Year, mexico_line.LEB, color='purple')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('LEB')
plt.title('Mexico')

ax = plt.subplot(2, 3, 5)
plt.plot(usa_line.Year, usa_line.LEB, color='green')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('LEB')
plt.title('USA')

ax = plt.subplot(2, 3, 6)
plt.plot(zimbabwe_line.Year, zimbabwe_line.LEB, color='magenta')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('LEB')
plt.title('Zimbabwe')

plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.show()
```

```python
#Zimbabwe had the most significant increase in life expectancy in the 2000-2015 period; after decreasing a bit 
#between 2000-2005, the country improved LEB approximately 15 years between 2005-2015
```

```python
#Plotting a bar graph of the 2015 LEB
chile_2015_leb = df[(df.Country == 'Chile') & (df.Year == 2015)]
china_2015_leb = df[(df.Country == 'China') & (df.Year == 2015)]
germany_2015_leb = df[(df.Country == 'Germany') & (df.Year == 2015)]
mexico_2015_leb = df[(df.Country == 'Mexico') & (df.Year == 2015)]
usa_2015_leb = df[(df.Country == 'United States of America') & (df.Year == 2015)]
zimbabwe_2015_leb = df[(df.Country == 'Zimbabwe') & (df.Year == 2015)]

plt.figure(figsize=(10, 7))
ax = plt.subplot()
plt.bar(len(chile_2015_leb), chile_2015_leb.LEB, color='gold')
plt.bar(len(china_2015_leb)+1, china_2015_leb.LEB, color='red')
plt.bar(len(germany_2015_leb)+2, germany_2015_leb.LEB, color='blue')
plt.bar(len(mexico_2015_leb)+3, mexico_2015_leb.LEB, color='purple')
plt.bar(len(usa_2015_leb)+4, usa_2015_leb.LEB, color='green')
plt.bar(len(zimbabwe_2015_leb)+5, zimbabwe_2015_leb.LEB, color='magenta')
ax.set_xticks(range(1, 7))
ax.set_xticklabels(['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'])
ax.set_yticks(range(20, 90, 20))
plt.xlabel('Country')
plt.ylabel('LEB')
plt.title('2015 LEB')

plt.show()
```

```python
#In 2015, Germany had the highest LEB (81 years), closely followed by Chile. There were more than 20 
#years of difference between Germany and Zimbabwe that year
```

```python
#Making a function that calculates the mean of GDP for each country 
def gdp_mean(country, data):
    subset = data[data.Country == country]
    mean = np.mean(subset.GDP)
    return '{} GDP mean (years 2000-2015): {}'.format(country, mean)

chile_gdp_mean = gdp_mean('Chile', df)
china_gdp_mean = gdp_mean('China', df)
germany_gdp_mean = gdp_mean('Germany', df)
mexico_gdp_mean = gdp_mean('Mexico', df)
usa_gdp_mean = gdp_mean('United States of America', df)
zimbabwe_gdp_mean = gdp_mean('Zimbabwe', df)

print(chile_gdp_mean)
print(china_gdp_mean)
print(germany_gdp_mean)
print(mexico_gdp_mean)
print(usa_gdp_mean)
print(zimbabwe_gdp_mean)
```

```python
#Between years 2000 and 2015, USA had the highest GDP mean. Zimbabwe, again, is the last in the list
```

```python
#Plotting line graphs of the GDP evolution over 15 years for each country
plt.figure(figsize=(10, 7))
ax = plt.subplot()
plt.plot(chile_line.Year, chile_line.GDP, color='gold', label='Chile')
plt.plot(china_line.Year, china_line.GDP, color='red', label='China')
plt.plot(germany_line.Year, germany_line.GDP, color='blue', label='Germany')
plt.plot(mexico_line.Year, mexico_line.GDP, color='purple', label='Mexico')
plt.plot(usa_line.Year, usa_line.GDP, color='green', label='USA')
plt.plot(zimbabwe_line.Year, zimbabwe_line.GDP, color='magenta', label='Zimbabwe')
ax.set_xticks(range(2000, 2016)) 
ax.set_xticklabels(['2000', '2001', '2002', '2003', '2004', '2005', '2006', 
                   '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015'], rotation=30, fontsize=10)
plt.xlabel('Years')
plt.ylabel('GDP (trillions of US Dollars)')
plt.title('GDP evolution over 15 years')
plt.legend()
plt.show()

plt.figure(figsize=(17, 10))

ax = plt.subplot(2, 3, 1)
plt.plot(chile_line.Year, chile_line.GDP, color='gold')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('GDP (trillions of US Dollars)')
plt.title('Chile')

ax = plt.subplot(2, 3, 2)
plt.plot(china_line.Year, china_line.GDP, color='red')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('GDP (trillions of US Dollars)')
plt.title('China')

ax = plt.subplot(2, 3, 3)
plt.plot(germany_line.Year, germany_line.GDP, color='blue')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('GDP (trillions of US Dollars)')
plt.title('Germany')

ax = plt.subplot(2, 3, 4)
plt.plot(mexico_line.Year, mexico_line.GDP, color='purple')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('GDP (trillions of US Dollars)')
plt.title('Mexico')

ax = plt.subplot(2, 3, 5)
plt.plot(usa_line.Year, usa_line.GDP, color='green')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('GDP (trillions of US Dollars)')
plt.title('USA')

ax = plt.subplot(2, 3, 6)
plt.plot(zimbabwe_line.Year, zimbabwe_line.GDP, color='magenta')
ax.set_xticks([2000, 2005, 2010, 2015])
ax.set_xticklabels(years, fontsize=10)
plt.xlabel('Years')
plt.ylabel('GDP (trillions of US Dollars)')
plt.title('Zimbabwe')

plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.show()
```

```python
#USA and China had a significant increase in their GDP between years 2000 and 2015. It is worth noting that in the 
#2008-2009 period, every country analyzed here, except China, saw their GDP decrease a bit
```
